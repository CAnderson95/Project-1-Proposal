```python
obama_median_list_price = votes_clean_state_median_listing_price_df.drop('2016 Election Party', axis=1)
obama_median_list_price.head()
```

```python
#%Change Total:
obama_median_list_price['PctChg'] = (obama_median_list_price['2017-09'] - obama_median_list_price['2012-11'])/obama_median_list_price['2017-09'] * 100
obama_median_list_price.head()
```

```python
dem_obama_median_list_price = obama_median_list_price.loc[(obama_median_list_price["2012 Election Party"] == 'Dem')]
dem_obama_median_list_price.head()
```

```python
dem_mean_pctchg = dem_obama_median_list_price['PctChg'].mean()
dem_mean_pctchg
```

```python
rep_obama_median_list_price = obama_median_list_price.loc[(obama_median_list_price["2012 Election Party"] == 'Rep')]
rep_obama_median_list_price.head()
```

```python
rep_mean_pctchg = rep_obama_median_list_price['PctChg'].mean()
rep_mean_pctchg
```

```python
#bargraph of %Change during Obama Election by State
f, ax = plt.subplots(figsize=(70,20))
dem = plt.bar(dem_obama_median_list_price.index, dem_obama_median_list_price['PctChg'], color="b", align="center")
rep = plt.bar(rep_obama_median_list_price.index, rep_obama_median_list_price['PctChg'], color="r", align="center")
plt.title("Percent Change of Listing Price by State Party Vote over 11 Months Following election of Obama")
plt.xlabel("State")
plt.ylabel("Percent Change")
plt.legend((dem, rep), ('Democrat', 'Republican'))
plt.show()
```
